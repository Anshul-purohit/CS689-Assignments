```python
with open("random_text_EN.txt",'r',encoding="utf-8") as file:
    text = file.readlines()

with open("../../test.en",'r',encoding="utf-8") as file:
    text1 = file.readlines()
    
with open("../../test.hi",'r',encoding="utf-8") as file:
    text2 = file.readlines()
    
result = []

for i in range(0,len(text)):
    for j in range(0,len(text1)):
        if(text[i]==text1[j]):
            result.append(text2[j])
```

```python
for i in text:
    print(i)
```

```python
for i in result:
    print(i)
```
