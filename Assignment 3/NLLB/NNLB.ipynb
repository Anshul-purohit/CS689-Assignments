## Part 1

### 1. English to Hindi

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import random
```

```python
with open("/kaggle/input/languagedataset/test.en","r",encoding="utf-8") as file:
    text_EN = file.readlines()
    
random_text_EN = random.sample(text_EN,1000) 
```

```python
for i in random_text_EN:
    print(i)
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M", src_lang="eng_Latn"
)
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
```

```python
model = model.to("cuda")
```

```python
def translate_EN_to_HI(article):
    inputs = tokenizer(article, return_tensors="pt").to("cuda")

    translated_tokens = model.generate(
        **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["hin_Deva"], max_length=30
    )
    return tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
```

```python
en_to_hi = []
for i in random_text_EN:
    en_to_hi.append(translate_EN_to_HI(i))
```

```python
for i in en_to_hi:
    print(i)
```

```python
with open('random_text_EN.txt', 'w', encoding='utf-8') as file:
    for i in random_text_EN:
        file.writelines(i+'\n')

with open('En_to_Hi.txt', 'w', encoding='utf-8') as file:
    for i in en_to_hi:
        file.writelines(i+'\n')
```

```python
import nltk
from nltk.translate.bleu_score import corpus_bleu
from rouge import Rouge
```

```python
def bleu_score(ground_truth , output):
    b_score = 0
    b_score = corpus_bleu([[ref] for ref in ground_truth], output)
    return b_score
   
def rouge_score(ground_truth , output):
    rouge = Rouge()
    scores = rouge.get_scores(ground_truth,output,avg=True)
    return scores
```

### 1. English to Hindi

```python
with open("English_to_Hindi/En_to_Hi.txt","r",encoding="utf-8") as file:
    En_to_Hi = file.readlines()

with open("English_to_Hindi/En_to_Hi_Original.txt","r",encoding="utf-8") as file:
    En_to_Hi_Original = file.readlines()
```

```python
print("Bleu Score : ",bleu_score(En_to_Hi_Original,En_to_Hi))
```

```python
rougeScore = rouge_score(En_to_Hi_Original,En_to_Hi)

for i,j in rougeScore.items():
    print(i , " : " , j)               
```
