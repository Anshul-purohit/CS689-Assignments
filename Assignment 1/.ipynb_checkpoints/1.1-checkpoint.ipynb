```python
def search_array(arr, target):
    for element in arr:
        if element == target:
            return True
    return False


halant = '\u094D'
a = 'अ'

vowels = ['अ', 'आ', 'इ', 'ई', 'उ', 'ए', 'ऐ', 'ओ', 'औ']

consonants = ['क', 'ख', 'ग', 'घ', 'ङ', 'च', 'छ', 'ज', 'झ', 'ञ', 'ट', 'ठ', 'ड', 'ढ', 'ण', 'त', 'थ', 'द', 'ध', 'न', 'प', 'फ', 'ब', 'भ', 'म', 'य', 'र', 'ल', 'व', 'श', 'ष', 'स', 'ह']

matra_to_vowels = {
    'ा': 'आ',   # a matra
    'ि': 'इ',   # i matra
    'ी': 'ई',   # ii matra
    'ु': 'उ',   # u matra
    'ू': 'ऊ',   # uu matra
    'ृ': 'ऋ',   # ri matra
    'ॄ': 'ऋ',   # rii matra
    'ॅ': 'ऌ',   # vocalic l matra
    'ॆ': 'ए',   # e matra
    'े': 'ए',   # ee matra
    'ै': 'ऐ',   # ai matra
    'ो': 'ओ',   # o matra
    'ौ': 'औ',   # au matra
    'ं': 'अं',   # chandrabindu matra 
    'ः': 'अः',   # anunasika matra
}

vowels_to_matras = {
    'अ' : '',
    'आ': '\u093E',  # आ
    'इ': '\u093F',  # इ
    'ई': '\u0940',  # ई
    'उ': '\u0941',  # उ
    'ऊ': '\u0942',  # ऊ
    'ए': '\u0947',  # ए
    'ऐ': '\u0948',  # ऐ
    'ओ': '\u094B',  # ओ
    'औ': '\u094C',  # औ
    'अं':  'ं',
    'अः': 'ः'
}

# with open('hi_100.txt', 'r', encoding='utf-8') as file:
#     # Read the entire contents of the file
#     str = file.read()


anusvara = 'ं'
anunasik = 'ँ'
ri = 'ऋ'


# for i in str:
#     print(i)

def something(str):
    ans=[]
    for i in str:
        if(i>='0' and i<='9'):
            ans.append(i)
            continue
        if(i==" "):
            ans.append(" ")
            continue
        if(i==anusvara):
            ans.append('अं')
            continue

        if(i==anunasik):
            if(len(ans)>0):
                temp = ans[len(ans)-1]
                ans = ans[:-1]
                ans.append(temp+i)
                continue

        if(i==halant):
            ans = ans[:-1]

        if(i==ri):
            ans.append(ri)
            continue

        find = search_array(vowels,i)
        if(find):
            ans.append(i)
            continue

        find = search_array(consonants,i)
        if(find):
            ans.append(i+halant)
            ans.append(a)

        for key in matra_to_vowels:
            if(i==key):
                ans = ans[:-1]
                ans.append(matra_to_vowels[key])
    return ans

text = 'मृत्युंजय'
ans = something(text)
print(ans)
# x = len(ans)

# for i in range(0,100):
#     print(ans[i])

# print(ans)
```

# Question : 2

```python
result = []
# ans = []

with open('hi_100.txt', 'r', encoding='utf-8') as file:
#     Read the entire contents of the file
    str = file.read()

# str = "आसमान में तारे चमक रहे हैं।"

token = str.split()
# print(token)

for word in token:
    result.append(something(word))

    
for i in range(0,10):
    print(result[i])
```

```python
# Unigram characters

frequency = {}

for sublist in result:
    for char in sublist:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1
            
sorted_freq = sorted(frequency.items(), key=lambda item: item[1], reverse=True)

final = sorted_freq[:20]

for i,j in final:

    print(i, " : ", j)
```

```python
# Bigram characters

biT = []

# ans = ['र्', 'आ', 'ष्', 'ट्', 'र्', 'ई', 'य्', 'अ', 'क्', 'ऐ', 'अं', 'स्', 'अ', 'र्', 'अ']

fresult = [char for sublist in result for char in sublist]

for i in range(1,len(fresult)-1):
    x = i
    y = i-1
    biT.append(fresult[y]+fresult[x])

# print(biT)

bi = {}
for i in biT:  
    if i in bi:
        bi[i] = bi[i] + 1
    else:
        bi[i] = 1
sorted_bi = sorted(bi.items(), key=lambda item: item[1], reverse=True)

finalb = sorted_bi[:20]

for i,j in finalb:
    print(i, " : ", j)
```

```python




# str = "आसमान में तारे चमक रहे हैं।"

# print(result)

# single_result = [item for sublist in result for item in sublist]


def syllable(s):
    i = 0
    res = []
    str = ""
    while i<len(s):
        if s[i] in vowels:
            str = str + s[i]
            if(i+1<len(s) and s[i+1]=='अं'):
                str = str + s[i+1]
                i = i + 1
            res.append(str)
            str = ""
        else:
            str = str + s[i]
        i = i + 1
    if(len(str)>0):
        res.append(str)

    ans = []
#     print(res)
    temp = ""
    for i in res:
        for j in i:
            if j in consonants:
                for k in consonants:
                    if(k==j):
                        temp = temp + k
                        break
                continue
            if (j==halant):
                temp = temp + j
                continue
            if j in vowels:
                if(len(temp)==0):
                    temp = temp + j
                    continue
                xx = temp[len(temp)-1]
                if(xx==halant):
                    temp = temp[:-1]
                    temp = temp + vowels_to_matras[j]
                else:
                    temp = temp + vowels_to_matras[j]
            else:
                if(len(temp)>0):
                    temp = temp + j
        if(temp):
            ans.append(temp)
        temp = ""

    return ans

# xx = []
# ss = syllable(single_result)
```

```python
# Unigram syllables

ss = syllable(fresult)
uni = {}
for i in ss:
    if i in uni:
        uni[i] = uni[i] + 1
    else:
        uni[i] = 1

sorted_unii = sorted(uni.items(), key=lambda item: item[1], reverse=True)   
# print(sorted_uni)

finaluu = sorted_unii[:20]

for i,j in finaluu:
    print(i, " : ", j)
```

```python
# Bigram syllables

biT = []

for i in range(1,len(ss)-1):
    x = i
    y = i-1
    biT.append(ss[y]+ss[x])

bi = {}
for i in biT:
    if i in bi:
        bi[i] = bi[i] + 1
    else:
        bi[i] = 1
sorted_bii = sorted(bi.items(), key=lambda item: item[1], reverse=True)
# print(sorted_bi)

finalbb = sorted_bii[:20]

for i,j in finalbb:
    print(i, " : ", j)
```
